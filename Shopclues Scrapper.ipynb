```python
import requests
from bs4 import BeautifulSoup
from IPython.display import Image,display
url="https://www.shopclues.com/search"
params={
    "q":"watch"
}
r=requests.get(url,params=params)
r.url
```

```python
soup=BeautifulSoup(r.content)
```

```python
products=soup.findAll('div',attrs={"class":"prd_p_section"})
images=soup.findAll('div',attrs={"class":"img_section"})
```

```python
k=1
for product in products:
    print(k)
    k=k+1
    for i in images:
        imgtag=i.find('img')
        display(Image(imgtag.attrs['data-img']))
        break
    prd_discount=product.find('span',attrs={"class":"prd_discount"}).text.strip()
    print(prd_discount)
    p_price=product.find('span',attrs={"class":"p_price"}).text.strip()
    print(p_price)
    print()

    
```
